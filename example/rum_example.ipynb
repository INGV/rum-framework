# RUM Example Notebook (Simplified)

This notebook simulates the RUM workflow: ingestion of raw files, generation of WF-Catalog, WF-Handle, and WF-Provenance with simulated PIDs.
All operations are local, no dependency on the real RUM repositories.

```python
# Imports
import json
import random
import string
from pathlib import Path
import pandas as pd
```

```python
# Helper function to generate random PIDs
def generate_pid():
    return "pid:" + ''.join(random.choices(string.ascii_letters + string.digits, k=8))
```

## Load dummy raw file

```python
# Example CSV dummy
raw_file = Path('example/raw_data.csv') # very fake data file
df = pd.read_csv(raw_file)
print(df.head())
```

## Simulate WF-Catalog

```python
wf_catalog = {
    "catalog_id": generate_pid(),
    "description": "Simulated WF-Catalog",
    "entries": df.to_dict(orient="records")
}

Path("json_examples").mkdir(exist_ok=True)
with open("json_examples/WF-Catalog_example.json", "w") as f:
    json.dump(wf_catalog, f, indent=2)
```

## Simulate WF-Handle

```python
wf_handle = {
    "handle_id": generate_pid(),
    "catalog_ref": wf_catalog["catalog_id"],
    "metadata": {"creator": "Simulated User", "type": "seismic_data"}
}

with open("json_examples/WF-Handle_example.json", "w") as f:
    json.dump(wf_handle, f, indent=2)
```

## Simulate WF-Provenance

```python
wf_prov = {
    "provenance_id": generate_pid(),
    "handle_ref": wf_handle["handle_id"],
    "source_file": str(raw_file)
}

with open("json_examples/WF-Provenance_example.json", "w") as f:
    json.dump(wf_prov, f, indent=2)
```

## Print generated JSONs

```python
print("WF-Catalog:")
print(json.dumps(wf_catalog, indent=2))

print("\nWF-Handle:")
print(json.dumps(wf_handle, indent=2))

print("\nWF-Provenance:")
print(json.dumps(wf_prov, indent=2))
```
