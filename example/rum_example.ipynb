# RUM Example Notebook (Simplified)

This notebook simulates the RUM workflow: ingestion of raw files, generation of WF-Handle, and WF-Provenance with simulated PIDs.
All operations are local, no dependency on the real RUM repositories.

```python
# Imports
import json
import random
import string
from pathlib import Path
import pandas as pd
```

```python
# Helper function to generate random PIDs
def generate_pid():
    return "prefix/" + ''.join(random.choices(string.ascii_letters + string.digits, k=8))
```

## Load dummy raw file

```python
# Example CSV dummy
raw_file = Path('/content/sample_data/anscombe.json') # very fake data file
df = pd.read_csv(raw_file)
print(df.head())
```

## PID generation

```python
# create PID
Pid = generate_pid() # one pid multiple json document
print('Pid =' + Pid)
```

## Simulate WF-Handle

```python
wf_handle = {
    "@context": {
                "dc": "http://purl.org/dc/elements/1.1/",
                "dcterms": "http://purl.org/dc/terms/",
                "schema": "http://schema.org/",
                "file": "http://schema.org/DigitalDocument"
            },
    "dc_identifier": Pid,
    "dc_title": "Example mSEED Repository",
    "dc_creator": "Example Institution",
    "dc_type": "Dataset, seismic waveform"
}

with open("json_examples/WF-Handle_example.json", "w") as f:
    json.dump(wf_handle, f, indent=2)
```

## Simulate WF-Provenance

```python
wf_prov = {
    "@context": {
                "dc": "http://purl.org/dc/elements/1.1/",
                "dcterms": "http://purl.org/dc/terms/",
                "schema": "http://schema.org/",
                "file": "http://schema.org/DigitalDocument"
            },
    "dc_identifier": Pid,
    "dcterms:isPartOf": "wf_handle",
    "prov_SoftwareAgent": "https://github.com/obspy/obspy",
    "schema_file": str(raw_file)
}

with open("json_examples/WF-Provenance_example.json", "w") as f:
    json.dump(wf_prov, f, indent=2)
```

## Print generated JSONs

```python
print("result RUM simulation")

print("\nWF-Handle:")
print(json.dumps(wf_handle, indent=2))

print("\nWF-Provenance:")
print(json.dumps(wf_prov, indent=2))
```
